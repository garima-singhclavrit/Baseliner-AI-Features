## **Gpt-2 tokenizer & pre-trained model**

```python
import tensorflow as tf
from transformers import GPT2Tokenizer, TFGPT2Model
```

```python
class GPT:
    def __init__(self):
        self.tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
        self.model = TFGPT2Model.from_pretrained('gpt2')

    def inp_vector(self, inp):
        encoded_inp = self.tokenizer(inp, return_tensors='tf')
        out = self.model(encoded_inp)
        vector = out[0].numpy()[0]
        return np.mean(vector, axis=0)

    def vectors(self, inps):
        feat_vectors = []
        for i in inps:
            if type(i) != 'str':
                i = "NA"
            feat_vectors.append(self.inp_vector(i))
        return np.array(feat_vectors)

gpt_model = GPT()
```

## **Define NN Architecture**

```python
# Branch-1 Task name
inp1 = tf.keras.Input(shape=(768,), dtype=tf.int32, name="input_layer_1")
lay1 = tf.keras.layers.Dense(128, activation='relu')(inp1)
lay1 = tf.keras.layers.Dropout(0.2)(lay1)
lay1 = tf.keras.layers.Dense(128, activation='relu')(lay1)

# Branch-2 Task description
inp2 = tf.keras.Input(shape=(768,), dtype=tf.int32, name="input_layer_2")
lay2 = tf.keras.layers.Dense(512, activation='relu')(inp2)
lay2 = tf.keras.layers.Dense(128, activation='relu')(lay2)

# Branch-2 Task label
inp3 = tf.keras.Input(shape=(768,), dtype=tf.int32, name="input_layer_3")
lay3 = tf.keras.layers.Dense(512, activation='relu')(inp3)
lay3 = tf.keras.layers.Dense(128, activation='relu')(lay3)

# Branch-2 Planned estimate
inp4 = tf.keras.Input(shape=(768,), dtype=tf.int32, name="input_layer_4")
lay4 = tf.keras.layers.Dense(512, activation='relu')(inp4)
lay4 = tf.keras.layers.Dense(128, activation='relu')(lay4)

# Branch-5 Final estimate
inp5 = tf.keras.Input(shape=(768,), dtype=tf.int32, name="input_layer_5")
lay5 = tf.keras.layers.Dense(512, activation='relu')(inp5)
lay5 = tf.keras.layers.Dense(128, activation='relu')(lay5)

# Concatenate Branch
out = tf.keras.layers.Concatenate()([lay1, lay2, lay3, lay4, lay5])
out = tf.keras.layers.Dense(384, activation='relu')(out)
out = tf.keras.layers.Dense(128, activation='relu')(out)
out = tf.keras.layers.Dense(216, activation='softmax')(out)

model = tf.keras.models.Model(inputs=[inp1, inp2, inp3, inp4, inp5], outputs=out)
```

```python
model.compile(tf.keras.optimizers.Adam(learning_rate=1e-3), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
checkpoint = tf.keras.callbacks.ModelCheckpoint('model.h5', monitor='loss', save_best_only=True, verbose=1)
```

```python
optest = []
for index, row in df_inp.iterrows():
  if 'MULE' or 'TISTUD' or 'APSTUD' or 'USERGRID' or 'JSW' or 'TDQ' in row['issuekey']:
    row["Optimistic Estimate"] = row['storypoint'] + round(random.uniform(1, 2))
  if 'BAM' or 'CLOV' or 'DURACLOUD' or 'MESOS' or 'TIMOB' in row['issuekey']:
    row["Optimistic Estimate"] = row['storypoint'] + round(random.uniform(2, 3))
  if 'DM' or 'MDL' or 'STUDIO' or 'XD' or 'TESB' in row['issuekey']:
    row["Optimistic Estimate"] = row['storypoint'] + round(random.uniform(3, 4))
  optest.append(row["Optimistic Estimate"])

df_inp["Optimistic Estimate"] = optest
```

```python
moslikest = []
pesest = []
for index, row in df_inp.iterrows():
  if row["storypoint"] > 6:
    row["Most Likely Estimate"] = round(row['Optimistic Estimate'] * round(random.uniform(1, 2), 1))
    row["Pessimistic Estimate"] = round(row['Optimistic Estimate'] * round(random.uniform(2, 3), 1))

  if row["storypoint"] <= 6:
    row["Most Likely Estimate"] = round(row['Optimistic Estimate'] * round(random.uniform(1, 2), 1))
    row["Pessimistic Estimate"] = row["Most Likely Estimate"]

  moslikest.append(row["Most Likely Estimate"])
  pesest.append(row["Pessimistic Estimate"])

df_inp["Most Likely Estimate"] = moslikest
df_inp["Pessimistic Estimate"] = pesest
```

## Neural Network

```python
# Branch-1 Issue key
inp1 = tf.keras.Input(shape=(768,), dtype=tf.int32, name="input_layer_1")
lay1 = tf.keras.layers.Dense(128, activation='relu')(inp1)
lay1 = tf.keras.layers.Dropout(0.2)(lay1)
lay1 = tf.keras.layers.Dense(128, activation='relu')(lay1)
```

```python
# Branch-2 Title
inp2 = tf.keras.Input(shape=(768,), dtype=tf.int32, name="input_layer_2")
lay2 = tf.keras.layers.Dense(512, activation='relu')(inp2)
lay2 = tf.keras.layers.Dense(128, activation='relu')(lay2)
```

```python
# Branch-3 Description
inp3 = tf.keras.Input(shape=(768,), dtype=tf.int32, name="input_layer_3")
lay3 = tf.keras.layers.Dense(512, activation='relu')(inp3)
lay3 = tf.keras.layers.Dense(128, activation='relu')(lay3)
```

```python
# Branch-4 Work Complexity
inp4 = tf.keras.Input(shape=(768,), dtype=tf.int32, name="input_layer_4")
lay4 = tf.keras.layers.Dense(128, activation='relu')(inp4)
lay4 = tf.keras.layers.Dropout(0.2)(lay4)
lay4 = tf.keras.layers.Dense(128, activation='relu')(lay4)
```

```python
# Branch-5 Actual Estimates
inp5 = tf.keras.Input(shape=(768,), dtype=tf.int32, name="input_layer_5")
lay5 = tf.keras.layers.Dense(128, activation='relu')(inp5)
lay5 = tf.keras.layers.Dropout(0.2)(lay5)
lay5 = tf.keras.layers.Dense(128, activation='relu')(lay5)
```

```python
# Branch-6 Risk Factor
inp6 = tf.keras.Input(shape=(768,), dtype=tf.int32, name="input_layer_6")
lay6 = tf.keras.layers.Dense(128, activation='relu')(inp6)
lay6 = tf.keras.layers.Dropout(0.2)(lay6)
lay6 = tf.keras.layers.Dense(128, activation='relu')(lay6)
```

```python
# Branch-7 High Confidence
inp7 = tf.keras.Input(shape=(768,), dtype=tf.int32, name="input_layer_7")
lay7 = tf.keras.layers.Dense(128, activation='relu')(inp7)
lay7 = tf.keras.layers.Dropout(0.2)(lay7)
lay7 = tf.keras.layers.Dense(128, activation='relu')(lay7)
```

```python
# Concatenate Branch
out = tf.keras.layers.Concatenate()([lay1, lay2, lay3, lay4, lay5, lay6, lay7])
out = tf.keras.layers.Dense(384, activation='relu')(out)
out = tf.keras.layers.Dense(128, activation='relu')(out)
out = tf.keras.layers.Dense(216, activation='softmax')(out)
```

```python
model = tf.keras.models.Model(inputs=[inp1, inp2, inp3, inp4, inp5, inp6, inp7], outputs=out)
```

## Model Compile

```python
model.compile(tf.keras.optimizers.Adam(learning_rate=1e-3), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
checkpoint = tf.keras.callbacks.ModelCheckpoint('model.h5', monitor='loss', save_best_only=True, verbose=1)

train_sh = model.fit(
    [issuekey_vec_train, title_vec_train, description_vec_train, work_comp_train, opt_est_train, risk_fac_train, high_conf_train], train_one_hot,
    epochs=100,
    callbacks=[checkpoint],
    validation_data=([issuekey_vec_test, title_vec_test, description_vec_test, work_comp_test, opt_est_test, risk_fac_test, high_conf_test], test_one_hot),
    batch_size=32,
    verbose=1
)
```
